    Plot part:

```
import numpy as np
import matplotlib.pyplot as plt

# Y-axis is volatge, we first input the voltage data for all the LEDs
voltage_blue = np.array([3.44, 2.65, 2.34, 1.7, 0.687, 0.131])
voltage_green = np.array([3.29, 2.61, 2.3, 1.667, 0.682, 0.131])
voltage_yellow = np.array([2.19, 1.74, 1.53, 1.11, 0.45, 0.123])

# Create the x-axis with pin values used
pin_value = np.array([255, 200, 175, 125, 50, 20])

# Set up correct title and labels
plt.xlabel('Pin Value')
plt.ylabel('Voltage')
plt.title('Voltage vs. Pin Value for LEDs')

# Plot the scatter plot for the data
plt.plot(pin_value, voltage_blue, '.', color = 'blue', label = 'Blue LED')
plt.plot(pin_value, voltage_green, '.', color = 'green', label = 'Green LED')
plt.plot(pin_value, voltage_yellow, '.', color = 'darkgoldenrod', label = 'Yellow LED')

# Calculate the best fit coefficients for the three LEDs
fit_blue = np.polyfit(pin_value, voltage_blue, deg = 1)
fit_green = np.polyfit(pin_value, voltage_green, deg = 1)
fit_yellow = np.polyfit(pin_value, voltage_yellow, deg = 1)

# Plot the best fit line for all three
plt.plot(pin_value, fit_blue[1] + fit_blue[0]* pin_value, '-', color = 'cornflowerblue', label = 'bestfit Blue LED')
plt.plot(pin_value, fit_green[1] + fit_green[0]* pin_value, '-', color = 'springgreen', label =  'bestfit Green LED')
plt.plot(pin_value, fit_yellow[1] + fit_yellow[0]* pin_value, '-', color = 'gold', label = 'bestfit Yellow LED')

plt.legend()

```

```
# Y-axis is the distances
y_axis = np.linspace(10, 100, 10)

# We also need to adjust the unit into meters, also, the actual distance the sound wave 
# travelled is 2 times of the one on the ruler. 
y_axis = y_axis*2/100
read_in_metrix = np.loadtxt('drive/My Drive/Classes/Physics 4BL/week1/duration1.txt' )
print(read_in_metrix) # check if the data is loaded correctly
```

```
# Calculate the mean duration time for each distance
mean_10 = np.mean(read_in_metrix[:,0])
mean_20 = np.mean(read_in_metrix[:,1])
mean_30 = np.mean(read_in_metrix[:,2])
mean_40 = np.mean(read_in_metrix[:,3])
mean_50 = np.mean(read_in_metrix[:,4])
mean_60 = np.mean(read_in_metrix[:,5])
mean_70 = np.mean(read_in_metrix[:,6])
mean_80 = np.mean(read_in_metrix[:,7])
mean_90 = np.mean(read_in_metrix[:,8])
mean_100 = np.mean(read_in_metrix[:,9])

x_axis = np.array([mean_10, mean_20, mean_30, mean_40, mean_50, mean_60, mean_70, mean_80, mean_90, mean_100])
# We also need to adjust the unit into seconds
x_axis = x_axis/1000000
print(x_axis)
```

```
# Plot the scatter plot of the means vs. time
plt.plot(x_axis, y_axis, '.')

# Set up correct title and labels
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.title('Time-of-flight vs. Distance')

slope, b = np.polyfit(x_axis, y_axis, deg = 1)
plt.plot(x_axis, b + slope * x_axis, '-')
print(slope)
```
